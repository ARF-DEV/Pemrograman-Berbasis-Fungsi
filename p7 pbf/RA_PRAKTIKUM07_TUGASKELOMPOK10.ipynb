```python
import pandas as pd
from functools import reduce
```

### 1. Model Populasi Polisi Detroit
Baca data dari file Laporan Polisi Detroit menggunakan modul csv dan terjemahkan data ke dalam daftar dictionary. Menggunakan Filter dengan fungsi lambda untuk mengecualikan dictionary (baris CSV) yang memiliki data yang hilang di kolom Zip, atau kolom Neighborhood. Dengan menggunakan fungsi lambda dan Reduce, hitung total waktu respons rata-rata, waktu pengiriman rata-rata, dan total waktu rata-rata untuk kepolisian Detroit.

```python
df = pd.read_csv("911_Calls_for_Service_(Last_30_Days).csv")
df = df.dropna()
df.columns
```

```python
dikti = []

for val in df.values:
  new_dict = {'zip': val[5], 'neighbor': val[20], 'total_response_time': val[17], 'total_time': val[19]}
  dikti.append(new_dict)
len(dikti)
```

```python
dikti_filtered = list(filter(lambda x: x['neighbor'] == x['neighbor'] and x['zip'] == x['zip'], dikti))
# nyari rata rata
mean_response = reduce(lambda a, b: a+b, df['totalresponsetime']) / len(df)
mean_delivery_time = reduce(lambda a, b: a+b, df['totalresponsetime']) / len(df)
dispatch_mean = reduce(lambda a, b: a+b, df['dispatchtime']) / len(df)
total_time_mean = reduce(lambda a, b: a+b, df['totaltime']) / len(df)

json_dict = {
    'mean_response': mean_response,
    'mean_delivery_time': mean_delivery_time,
    'dispatch_mean': dispatch_mean,
    'total_time_mean': total_time_mean
}
```

### 2. Modelkan Neighborhood Samples
Menggunakan fungsi lambda dan map, atau lambda dan Filter, bagi list dictionary menjadi list dictionary yang lebih kecil yang dipisahkan oleh neighborhood. Dengan menggunakan lambda dan Reduce, temukan total waktu respons rata-rata untuk setiap neighborhood, waktu pengiriman rata-rata untuk setiap neighborhood, dan total waktu rata-rata untuk setiap neighborhood dan simpanlah ke dalam list dictionary. Tambahkan item dictioanary untuk menyertakan data populasi untuk semua Detroit dalam daftar gabungan Anda.

```python
data = pd.read_csv('911_Calls_for_Service_(Last_30_Days).csv')

data.head()
```

```python
data.keys()
```

```python
# make empty list
neighborhoods = []

# data cleaning
data = data.dropna(subset=['neighborhood'])
data = data.dropna() #drop all column that na

# get all unique neighborhoods
for neighborhood in data['neighborhood']:
    if neighborhood not in neighborhoods:
        neighborhoods.append(neighborhood)
print(neighborhoods)
```

```python
# make a new dict for each neighborhood
neighborhood_dict = {}
for neighborhood in neighborhoods:
    neighborhood_dict[neighborhood] = []

# add data to each neighborhood dict
for _, row in data.iterrows():
    neighborhood_dict[row['neighborhood']].append(row.drop('neighborhood').to_dict())
```

```python
neighborhood_dict_summary = []

for neighborhood in neighborhoods:
    total_dispatch_time = reduce(lambda x, y: x + y, [row['dispatchtime'] for row in neighborhood_dict[neighborhood]])
    avg_dispatch_time = total_dispatch_time / len(neighborhood_dict[neighborhood])

    total_response_time = reduce(lambda x, y: x + y, [row['totalresponsetime'] for row in neighborhood_dict[neighborhood]])
    avg_response_time = total_response_time / len(neighborhood_dict[neighborhood])

    total_total_time = reduce(lambda x, y: x + y, [row['totaltime'] for row in neighborhood_dict[neighborhood]])
    avg_total_time = total_total_time / len(neighborhood_dict[neighborhood])

    neighborhood_dict_summary.append({
        'neighborhood': neighborhood,
        'avg_response_time': avg_response_time,
        'avg_dispatch_time': avg_dispatch_time,
        'avg_total_time': avg_total_time,
    })

neighborhood_dict_summary
```

### 3. Export to json
